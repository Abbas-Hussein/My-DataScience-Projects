### Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
x = np.linspace(0,10,100)
sin_x = np.sin(x)
cos_x= np.cos(x)
```

```python
plt.figure(figsize=(5,3))

plt.plot(x,sin_x,label="sin(x)",color='black')
plt.plot(x, cos_x,label='cos(x)',color='green') # Plot x against the cos of x

plt.xlabel("distance")
plt.ylabel(r"$\frac{distance}{time}$") # LateX code, r means raw act like f string and doesnt consider the python feature of \ as a new line
plt.legend(['sin(x)','cos(x)'])
plt.title("Sine wave graph")
```

```python
x = [0,1,2,3,4]
y = [0,0,1,0,0]

plt.plot(x,y,marker='o',markerfacecolor='black',markeredgecolor='green',markersize=10,linewidth=1,linestyle='-.') # More styles in matplotlib
```

```python
x = np.linspace(0,10,10)
sin_x = np.sin(x)
cos_x= np.cos(x)
tan_x = np.tan(x)
```

```python
plt.figure(figsize=(5,3))

plt.plot(x,sin_x,label="sin(x)",color='black',marker='o',markerfacecolor='white')
plt.plot(x, cos_x,label='cos(x)',color='green',marker='*',mfc='white') # Plot x against the cos of x
plt.plot(x,tan_x, label = "tan(x)",color='red',marker='^',mfc='white')
plt.xlabel("distance")
plt.ylabel(r"$\frac{distance}{time}$") # LateX code, r means raw act like f string and doesnt consider the python feature of \ as a new line
plt.legend()
plt.title("Sine wave graph")
```

### Barcharts

- They display categorical data using rectangular bars proportional to their values.

```python
gender = ['Male','Female']
count = [90, 92]
plt.figure(figsize=(8,3))
plt.bar(gender, count)
# plt.yticks(range(0,110,10)); # Used to set the y-axis values
plt.ylim([89,93]) # Limits the y-axis to the range between the first and second value
```

```python
plt.bar(gender, count, width=0.3, ec='blue',color='skyblue')
```

```python
plt.barh(gender,count,height=0.3,color=['lavender','palegreen'],edgecolor=['purple','forestgreen'])
```

### GROUPED BAR CHARTS

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
groups = ['Group 1','Group 2','Group 3','Group 4','Group 5']
MALES = [15,25,20,30,10]
FEMALES = [30,10,20,15,5]
bar_width = 0.35
indices = np.arange(len(groups))
```

```python
indices # This is for positioning the groups
```

```python
plt.bar(indices - bar_width / 2, MALES, width=bar_width,label="Male")
plt.bar(indices + bar_width / 2, FEMALES, width=bar_width,label="Female")
plt.xticks(indices,groups) # Telling matplotlib to name each group with the group list
plt.legend()
```

```python
plt.barh(indices - bar_width / 2, MALES, height=bar_width,label="Male")
plt.barh(indices + bar_width / 2, FEMALES, height=bar_width,label="Female")
plt.yticks(indices,groups) # Telling matplotlib to name each group with the group list
plt.legend()
```
