### Categorical Plots
We'll be concerned about seeing the distribution of the categorical column such as the sex(gender) column, smoker, day, time
in reference either to one of the numerical columns or another categorical col


```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Bar Plots: This is a general plot that allows you to aggregate the categorical data based of of some functions and by default (The mean)
sns.barplot(x='sex',y='total_bill',data=tips) # A plot between a numerical and categorical data
plt.show()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # estimator is a function for aggregate that is the mean but you can also 
plt.show()                                                               #do your own function, so here we used standard deviation
```

```python
# Count plot: same as the barplot but here the estimator only counts the number of occurences, hence we only set the x value)
sns.countplot(x='sex',data=tips) # SO in countplot, the y axis is automatically chosen for you
plt.show()
```

```python
# Box Plots: Box and violin plots are used to show the distribution of categorical data
#A box plot is sometimes known as a box or whisker plot and its just a distribution that shows comparism of variables
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show() # the dots above the whiskers are outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') # this enables us to split the boxplots even further
plt.show()                                                 # Now, we see the plots of both the day and smokers side by side
# So with this, you not only compare the day by bay behavior for total_bill but you also know who are smokers
```

```python
# Violin Plots
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show() # So violin plots are definitely related to each other but violin has more info
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split = True) # violin also has hue and can split both x and hue in one violin using split
plt.show()
```

```python
# Stripplot: It will draw a scatter plot where one variable is categorical 
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# The problem here is that we cant say how many points are stacked ontop of each other 
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True) # To do this, we use jitter to make it a bit better
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)# We can also add hue and split(use dodge for split here)
plt.show()
```

```python
# Swarm plot: This is when you combine stripplot and a violin plot. the points are adjusted so they dont overlap
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()      # Don't use this swarmplot often as people are not that acquanited with it
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black') # so you can literally join violin plots and swarmplot
plt.show()
```

```python
#Catplot(previously factorplot): This is most general and you just plot and choose the kind of chart you choose
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
plt.show()
```

### Matrix Plots


```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights =sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# HeatMaps: This is a primary way of showing some sort of matrix plots. To use this, we must put it in matrix form
#Correlation data (corr)
tc= tips.corr(numeric_only=True) #add numeric only
```

```python
tc
```

```python
sns.heatmap(tc)
plt.show()
```

```python
sns.heatmap(tc,annot=True) #So it annotates the numerical that belong to each of the cell
plt.show()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm') #cmap is for color map
plt.show()
```

```python
flights
```

```python
# To convert this to matrix form i.e i will convert one of the 3 columns to be the row name
flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp)
plt.show()
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=3)
plt.show()
```

```python
sns.heatmap(fp)
plt.show()
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
plt.show()
```

### Grids

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
# WE can use PairGrid
sns.PairGrid(iris)
plt.show() # This gives an empty grid
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)  # map a scatter plot
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
plt.show()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
plt.show()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
plt.show()
```

### Regression Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='sex') # You can still have hue here, remove row to reduce info
plt.show()
```
