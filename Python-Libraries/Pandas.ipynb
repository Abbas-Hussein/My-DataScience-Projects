# Types of data (statistics)
- Categorical (Qualitative)
 - Nominal: This type of data doesn't have an order or ranking. Each category have equal precedence. E.g. (sex, favorite color)
 - Ordinal: This type of data has order or ranking. E.g. (Position, Rating)
- Numerical (Quantitative)
 - Metric: This type of data can take values of any range. They are usually measurements.
 - Discrete: This type of data is countable

```python
# Query is used to filter and get some particular data
df.query('Age > 30 and Fare < 25') # Gets the rows with Age greater than 30 and Fare less than 25
```

```python
min_age = 30
max_fare = 25
df.query('Age > @min_age and Fare < @max_fare') # The '@' symbol is used before specifying a variable name. In this case it will replace @min_age with 30 and @max_fare with 25
```
